````python
alpha = 0.05  # 5% significance level
from scipy.stats import ttest_ind

# Filter out platforms
xone_ratings = relevant_years[(relevant_years['platform'] == 'XOne') & (relevant_years['user_score'].notnull())]['user_score']
pc_ratings = relevant_years[(relevant_years['platform'] == 'PC') & (relevant_years['user_score'].notnull())]['user_score']

# Run t-test (equal_var=False assumes unequal variance)
t_stat, p_value = ttest_ind(xone_ratings, pc_ratings, equal_var=False)

print(f"T-statistic: {t_stat:.3f}, P-value: {p_value:.3f}")

if p_value < alpha:
    print("‚ùå Reject the null hypothesis: average ratings are significantly different.")
else:
    print("‚úÖ Fail to reject the null hypothesis: no significant difference.")


# Filter out genres
action_ratings = relevant_years[(relevant_years['genre'] == 'Action') & (relevant_years['user_score'].notnull())]['user_score']
sports_ratings = relevant_years[(relevant_years['genre'] == 'Sports') & (relevant_years['user_score'].notnull())]['user_score']

# Run t-test
t_stat2, p_value2 = ttest_ind(action_ratings, sports_ratings, equal_var=False)

print(f"T-statistic: {t_stat2:.3f}, P-value: {p_value2:.3f}")

if p_value2 < alpha:
    print("‚ùå Reject the null hypothesis: Action and Sports ratings differ.")
else:
    print("‚úÖ Fail to reject the null hypothesis: no significant difference.")
````



### üìà Hypothesis Testing Summary

We conducted two hypothesis tests to evaluate differences in average user ratings between platforms and genres.

#### 1. **Xbox One vs. PC - User Ratings**

- **T-statistic**: 1.452  
- **P-value**: 0.148  
- **Conclusion**: ‚úÖ *Fail to reject the null hypothesis*

There is **no statistically significant difference** in average user ratings between **Xbox One and PC** platforms. The variation observed is likely due to random chance rather than a true difference in user perception.

---

#### 2. **Action vs. Sports Genres - User Ratings**

- **T-statistic**: 10.233  
- **P-value**: 0.000  
- **Conclusion**: ‚ùå *Reject the null hypothesis*

There is a **statistically significant difference** between the average user ratings for **Action** and **Sports** games. This suggests that players consistently rate these genres differently, potentially due to differences in gameplay quality, engagement, or expectations.
